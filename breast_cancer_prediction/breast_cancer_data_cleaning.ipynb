### BREAST CANCER DATASET

##### Introduction

Breast cancer diagnosis is a critical medical task, as misclassification especially false negatives which can lead to severe clinical consequences. This project aims to develop a robust machine learning pipeline to classify tumors as Benign (B) or Malignant (M) using the Breast Cancer dataset.

The primary objectives include:

- Cleaning and preprocessing the dataset

- Handling feature scaling and class imbalance

- Building and tuning Logistic Regression and Random Forest models

- Evaluating model performance using medically relevant metrics

- Selecting the best model based on recall and F1-score

#### Data cleaning: handle missing values

```python
# Create a copy of the dataset
breast_cancer_df = breast_cancer_data.copy()
```

```python
# Set the case id as the index
breast_cancer_df.set_index("id", inplace=True)
```

```python
# Display the first two rows
breast_cancer_df.head(2)
```

```python
# Handle the column with missing values (NAN)
breast_cancer_df = breast_cancer_df.drop('Unnamed: 32', axis=1)
```

```python
# Confirm missing colunmn has been dropped
breast_cancer_df.isna().sum()
```

```python
# Display the columns
breast_cancer_df.head(2)
```

```python
# Check the values in the target column (diagnosis)
breast_cancer_df['diagnosis'].unique()
```

#### Getting the data ready for EDA
- No duplicated values found.
- A column with invalid values was found and were handled (dropped).


```python
# Convert the diagnosis 
breast_cancer_df = breast_cancer_df.replace({"diagnosis":{'M':'Malignant', 'B': 'Benign'}})
```
